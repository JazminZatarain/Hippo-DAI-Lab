# Showing whether the optimizations converge

For this purpose, the 4 problem formulations from the directory `problemformulations` have been run. Each problem formulation used 100,000 nfe. The results and $\epsilon$-progress have been saved in the directory `results`.

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Loading convergence data

```python
df_con1 = pd.read_csv('results/NORDHAUS_UTILITARIAN_100000_convergence.csv')
df_con2 = pd.read_csv('results/NORDHAUS_SUFFICIENTARIAN_100000_convergence.csv')
df_con3 = pd.read_csv('results/WEITZMAN_UTILITARIAN_100000_convergence.csv')
df_con4 = pd.read_csv('results/WEITZMAN_SUFFICIENTARIAN_100000_convergence.csv')
```

## Setting labels

```python
# Set parameters
x_label = '$\epsilon$-progress'
y_label = 'nfe'
```

## Create subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2, figsize=(20, 10))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.5)
fig.patch.set_facecolor('white')

ax1.plot(df_con1.nfe, df_con1.epsilon_progress)
ax1.set_ylabel(x_label)
ax1.set_title('Nordhaus + Utilitarian')
ax1.set_xlabel(y_label)

ax2.plot(df_con3.nfe, df_con3.epsilon_progress)
ax2.set_ylabel(x_label)
ax2.set_title('Weitzman + Utilitarian')
ax2.set_xlabel(y_label)

ax3.plot(df_con2.nfe, df_con2.epsilon_progress)
ax3.set_ylabel(x_label)
ax3.set_title('Nordhaus + Sufficientarian')
ax3.set_xlabel(y_label)

ax4.plot(df_con4.nfe, df_con4.epsilon_progress)
ax4.set_ylabel(x_label)
ax4.set_title('Weitzman + Sufficientarian')
ax4.set_xlabel(y_label)

plt.show()
```

## Save figure

```python
fig.savefig("./results/convergence.png", dpi=400, pad_inches=0.2)
```
