# Visualizan of outcomes from optimization

```python
from ema_workbench.analysis import parcoords
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
directory = os.getcwd()
folder = '/results/'
file = 'WEITZMAN_UTILITARIAN_100000_results.csv'
```

```python
outcomes = pd.read_csv(directory + folder + file).iloc[:, 4:]

limits = parcoords.get_limits(outcomes)
axes = parcoords.ParallelAxes(limits)
axes.plot(outcomes)

plt.show()
```
